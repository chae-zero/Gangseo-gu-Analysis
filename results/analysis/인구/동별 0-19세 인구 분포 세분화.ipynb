# 시각화 세팅, 함수

```python
marker='o'
color='black'
title_size = 15
```

```python
def linegraph(df, title='제목 입력하기', rotation=0, path='./img/인구'):
    plt.plot(df, marker=marker, color=color)
    plt.xticks(rotation=rotation)
    plt.title(title, size=title_size)
    plt.tight_layout()
    plt.savefig(f'{path}/{title}.png')
    plt.show()
```

# 동별 0-19세 인구 분포 세분화

```python
for year in range(16, 23):
    path = f'../../data/인구/{year}년도 동별 연령별 인구.csv'
    title = f'동별 0-19세 인구 분포 세분화 20{year}'
    img_path = f'../../img/인구/동별 0-19세 인구 분포 세분화'
    df = pd.read_csv(path)

    plt.figure(figsize=(18,10))

    X = np.arange(0, 40, 2)
    bar_width = 0.40

    plt.bar(X + bar_width * 0,
            df['0~4세'] ,
            color='r',
            label='0~4세',
            width=bar_width)

    plt.bar(X + bar_width * 1,
            df['5~9세'] ,
            color='b',
            label='5~9세',
            width=bar_width)

    plt.bar(X + bar_width * 2,
            df['10~14세'] ,
            color='g',
            label='10~14세',
            width=bar_width)

    plt.bar(X + bar_width * 3,
            df['15~19세'],
            color='pink',
            label='15~19세',
            width=bar_width)

    plt.legend()
    plt.title(title, size=title_size)
    plt.xticks(np.arange(bar_width, 40 + bar_width, 2), df['동별'], rotation=90)
    plt.savefig(f'{img_path}/{title}_1.png')
    plt.show()


    plt.figure(figsize=(18,10))

    X = np.arange(0, 20)

    plt.bar(X,
            df['0~4세'] ,
            color='r',
            label='0~4세')

    plt.bar(X,
            df['5~9세'] ,
            color='b',
            label='5~9세',
            bottom=df['0~4세'])

    plt.bar(X,
            df['10~14세'] ,
            color='g',
            label='10~14세',
            bottom=df['0~4세'] + df['5~9세'])

    plt.bar(X,
            df['15~19세'],
            color='pink',
            label='15~19세',
            bottom=df['0~4세'] + df['5~9세'] + df['10~14세'])

    plt.legend()
    plt.title(title, size=title_size)
    plt.xticks(X, df['동별'], rotation=90)
    plt.savefig(f'{img_path}/{title}_2.png')
    plt.show()
```
