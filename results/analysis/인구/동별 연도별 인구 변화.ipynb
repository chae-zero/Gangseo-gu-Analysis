# 강서구 인구 연도별 추이 데이터 확인

```python
path = 'data/인구/강서구 동별 인구 추이 2016-2022.csv'
df = pd.read_csv(path)
df.tail(2)
```

```python
df2 = df.set_index('동별(3)', drop=True)
df2
```

# 시각화 세팅, 함수

```python
marker='o'
color='black'
title_size = 15
```

```python
def linegraph(df, title='제목 입력하기', rotation=0, path='./img/인구'):
    plt.plot(df, marker=marker, color=color)
    plt.xticks(rotation=rotation)
    plt.title(title, size=title_size)
    plt.tight_layout()
    plt.savefig(f'{path}/{title}.png')
    plt.show()
```

# 2016-2022년 강서구 인구수 변화

```python
linegraph(df2.loc['합계'], "강서구 인구수 변화 2016-2022")
```

# 2016-2022년 동별 인구수 변화

```python
for dong in df2.index[:-1]:
    linegraph(df2.loc[dong, :], f"{dong} 인구수 변화 2016-2022")
```
