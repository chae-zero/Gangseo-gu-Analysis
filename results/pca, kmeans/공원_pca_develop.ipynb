# PCA develop

```python
data_scaled = df_all.copy()

ss = StandardScaler()

for col in data_scaled.columns:
    data_scaled[col] = ss.fit_transform(data_scaled[[col]])
    
display(data_scaled.shape, data_scaled.head(3))
```

```python
variance_ratio = {}

for i in range(1, len(data_scaled.columns)+1):
    pca = PCA(n_components=i)
    pca.fit(data_scaled)
    variance_ratio[f'n_{i}'] = pca.explained_variance_ratio_.sum()
```

```python
variance_ratio
```

```python
pca = PCA(n_components=i)
pca.fit(data_scaled)
```

```python
# eigenvector per each PC
data_pca3 = pd.DataFrame(pca.components_[0:3],
                         columns=data_scaled.columns,
                         index = ['PC1','PC2','PC3']).T

data_pca3
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(data_pca3,
            annot=True,
            cmap='RdYlGn',
            cbar_kws={'shrink' : 0.5}           
           )

plt.show()
```

```python
# eigen value
eigen_value = np.sort(pca.explained_variance_)[::-1]

plt.figure(figsize=(12, 5))

plt.plot(variance_ratio.keys(), eigen_value)
plt.ylim(0, 10, 1)
plt.axhline(1, color = 'red', ls = '--')
plt.title('Elbow Point')

plt.show()
```

```python
plt.figure(figsize = (12, 5))

plt.plot(variance_ratio.keys(), variance_ratio.values())
plt.axhline(0.7, color = 'red', ls = '--', lw = 1)
plt.axhline(0.9, color = 'red', ls = '--', lw = 1.5)
plt.title("Variance Ratio")

plt.show()
```

```python
pca = PCA(n_components = 3, random_state = 42)

pca.fit(data_scaled)
data_pca = pd.DataFrame(pca.transform(data_scaled), 
                        columns = (["PC1", "PC2", "PC3"]))

data_pca.describe().T
```

```python
km = KMeans()
Elbow_M = KElbowVisualizer(estimator = km, k = 10, metric='silhouette')
Elbow_M.fit(data_pca)
Elbow_M.show()

plt.show()
```

```python
km = KMeans(n_clusters=3, random_state=42)

yhat_AC = km.fit_predict(data_pca)

data_pca["Clusters"] = yhat_AC   # for evaluating the model
```

```python
x =data_pca["PC1"]
y =data_pca["PC2"]
z =data_pca["PC3"]

fig = plt.figure(figsize=(18,10))
ax = plt.subplot(111, projection='3d')
ax.scatter(x, y, z, s=30, c=data_pca["Clusters"], marker='o', alpha = 0.5, cmap = 'jet')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title("The Plot Of The Clusters(3D)")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(data = data_pca, x = 'PC1', y='PC2', hue='Clusters')
plt.title('The Plot Of The Clusters(2D)')
plt.show()
```

```python
ax = sns.countplot(x=data_pca["Clusters"])
ax.bar_label(ax.containers[0])
ax.set_title("Distribution Of The Clusters")
plt.show()
```

# PCA X, Y, Z

```python

pca = PCA(n_components=1)
pca.fit(df_counting)
x_pca = pca.transform(df_counting)
```

```python
pca = PCA(n_components=1)
pca.fit(df_distance)
y_pca = pca.transform(df_distance)
```

```python
pca = PCA(n_components=1)
pca.fit(df_elder)
z_pca = pca.transform(df_elder)
```

```python
mean_v, std_v = x_pca.mean(), x_pca.std()
x_pca = (x_pca - mean_v) / std_v
```

```python
mean_v, std_v = y_pca.mean(), y_pca.std()
y_pca = (y_pca - mean_v) / std_v
```

```python
mean_v, std_v = z_pca.mean(), z_pca.std()
z_pca = (z_pca - mean_v) / std_v
```

# K-means elbow 시각화

```python
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(df_all)
    cs.append(kmeans.inertia_)

plt.figure(figsize=figure_size)
plt.plot(range(1, 11) , cs , 'o')
plt.plot(range(1, 11) , cs , '-' , alpha = 0.5)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters') , plt.ylabel('CS')
plt.savefig(f'../img/pca, kmeans/The Elbow Method.png')
plt.show()
```

# Agglomerative Hierarchical Clustering

```python
df_xy['PARK_SN'] = df_raw['PARK_SN']
```

```python
df_xy
```

```python
from sklearn.cluster import AgglomerativeClustering 

n = 3
agglom = AgglomerativeClustering(n_clusters=n, linkage='average').fit(df_xy)

df_xy['Labels'] = agglom.labels_
fig, ax = plt.subplots(figsize=figure_size)
ax.scatter(x=df_xy['건물용도별 개수'], y=df_xy['건물용도별 최단거리'], 
               )
for i, txt in enumerate(df_xy['PARK_SN']):
    ax.annotate(txt, (df_xy['건물용도별 개수'][i], df_xy['건물용도별 최단거리'][i]))
    
plt.title(f'Agglomerative with {n} Clusters', size=title_size)
plt.savefig(f'../img/pca, kmeans/건물용도별 최단거리와 고령인구 데이터 pca - Agglomerative with {n} Clusters.png')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 

n = 3
agglom = AgglomerativeClustering(n_clusters=n, linkage='average').fit(df_xy)

df_xy['Labels'] = agglom.labels_
plt.figure(figsize=figure_size)
sns.scatterplot(x='건물용도별 개수', y='건물용도별 최단거리', data=df_xy, hue=df_xy['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title(f'Agglomerative with {n} Clusters', size=title_size)
plt.savefig(f'../img/pca, kmeans/건물용도별 최단거리와 고령인구 데이터 pca - Agglomerative with {n} Clusters.png')
plt.show()
```

```python
from scipy.cluster import hierarchy 
from scipy.spatial import distance_matrix 
```

```python
df_xy.drop('Labels', axis=1, inplace=True)
```

```python
dist = distance_matrix(df_xy, df_xy)
print(np.round(dist, 10))
```

```python
Z = hierarchy.linkage(dist, 'complete')
```

```python
plt.figure(figsize=(18, 30))
dendro = hierarchy.dendrogram(Z, leaf_rotation=0, leaf_font_size=14, orientation='right')
plt.title(f'Dendrogram', size=title_size)
plt.xlabel('distance')
plt.ylabel('PARK_SN')
plt.savefig(f'../img/pca, kmeans/Dendrogram.png')
plt.show()
```
